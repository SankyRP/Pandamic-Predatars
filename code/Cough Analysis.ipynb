```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from glob import glob

import librosa as lr
```

```python
data_dir = 'D:\SomeShodhan\covid\data\cough\pos'
audio_files = glob(data_dir + '/*.wav')

len(audio_files)
```

```python
audio, sfreq = lr.load(audio_files[0])
time = np.arange(0, len(audio)) / sfreq

time
```

```python
fig, ax = plt.subplots()
ax.plot(time, audio)
ax.set(xlabel = 'Time (s)' , ylabel = 'Sound(Amplitude)')
plt.show()
```

```python
%%time

for file_no in range(0, len(audio_files), 1):
    
    audio, sfreq = lr.load(audio_files[file_no])
    time = np.arange(0, len(audio)) / sfreq
    
    fig, ax = plt.subplots()
    ax.plot(time, audio)
    ax.set(xlabel = 'Time (s)' , ylabel = 'Sound(Amplitude)')
    plt.show()

    
```
